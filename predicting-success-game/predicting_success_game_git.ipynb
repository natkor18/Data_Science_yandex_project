```python
data_max_sales_platform = data.pivot_table(index='platform', values = 'region_sum_sales', aggfunc='sum')
data_max_sales_platform = data_max_sales_platform.sort_values('region_sum_sales', ascending=False)
```

<font size="4">Лидирующее положение занимают шесть платформ: PS2, X360, PS3, Wii, DS, PS. Остальные значительно уступают. Поосмотрим для этих платформ распределение по годам. </font>

```python
#data.pivot_table(index='days_publication',values='last_price',aggfunc='mean').plot(y='last_price', kind='line', title='mean');
top6_platform = data_max_sales_platform[:6]
top6_platform
```

```python
plt.figure(figsize=(15, 10))
plt.title(f'Распределение суммарных продаж топ6 платформ по годам', fontsize=15)
for item in top6_platform.index:
    result = data.query('platform == @item').pivot_table(index='year_of_release', values='region_sum_sales', aggfunc='sum')
    plt.plot(result, 'o-', label=item, alpha=1)
    plt.title("Количество игр выпускаемые в разные года")
    plt.xlabel("Год выпуска")
    plt.ylabel("Количество выпущенных игр")
    plt.grid(True)
    plt.legend()
plt.show();
```

```python
for item in top6_platform.index:
    plt.figure(figsize=(5,5))
    plt.title(item, fontsize=15)
    result = data.query('platform == @item').pivot_table(index='year_of_release', values='region_sum_sales', aggfunc='sum')
    plt.plot(result, 'o-', label=item, alpha=1)

    plt.xlabel("Год выпуска")
    plt.ylabel("Количество выпущенных игр")
    plt.grid(True)
    plt.legend()
```

# Шаг 5. Проверьте гипотезы

**Гипотезы**
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

<font size="4">1) **Гипотеза:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;</font>

```python
XOne = data_actual[(data_actual['platform'] == 'XOne') &  (data_actual['user_score'] > 0)]['user_score']
PC = data_actual[(data_actual['platform'] == 'PC') &  (data_actual['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг платформы Xbox One:', round(XOne.mean(),2))
print('Средний пользовательский рейтинг платформы PC:', round(PC.mean(),2))
```

<font size="4">H0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1: Средние пользовательские рейтинги платформ Xbox One и PC разные</font>

```python
#проверяем первую гипотезу

alpha = 0.05

results = st.ttest_ind(XOne, PC, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<font size="4">**Вывод: Первая гипотеза подтвердилась. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.**</font>


<font size="4">2) **Гипотеза**: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

H0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые

H1: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные</font>

```python
action = data_actual[(data_actual['genre'] == 'Action') &  (data_actual['user_score'] > 0)]['user_score']
sports = data_actual[(data_actual['genre'] == 'Sports') &  (data_actual['user_score'] > 0)]['user_score']
print('Средний пользовательский рейтинг жанра Action:', action.mean())
print('Средний пользовательский рейтинг жанра Sports', sports.mean())
```

```python
#проверяем вторую гипотезу

alpha = 0.05

results = st.ttest_ind(action, sports, equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<font size="3">Использует метод st.ttest_ind, который проверяет средние двух совокупностей на равенство. Нулевая гипотеза  предполагает равенство средних исследуемых генеральных совокупностей. Полученное p-value показывает вероятность, с которой мы ошибемся, приняв альтернативную гипотезу. Если она достаточно мала, можем отвергнуть нулевую гипотезу и принять альтернативную (средние не равны), если достаточно велика - не можем отвергнуть нулевую, приходится её принять (средние равны).</font>

<font size="4">**Вывод: Вторая гипотеза не подтвердилась. Средние пользовательские рейтинги жанров Action и Sports разные**</font>

- Середина 90х -начало 2000х лидирует плафторма PS, с 2000 и вплоть до 2005 года на смену ей приходит PS2, в 2006 году появляется Wii и прорывается вверх. Примерно с 2010 года наблюдается обвал продаж для всех платформ.<br><br>

- Средний жизненный цикл платформы: 11 лет. Пик популярности приходится на 4-5 годы, затем наблюдается спад. В этот момент надо готовить новый релиз платформы.<br><br>
За период в три года на рынке осталось 10 самых популярных платформ - PS4, 3DS, XOne, WiiU, PS3, X360, PC, Wii, PSV, PSP<br><br>

- Лидирующие позиции занимают платформы PS4, XOne, WiiU, X360, Wii. Все они скошены вправо. У них длинные усы, платформы имею большее число продаж.<br><br>

- Наблюдается весьма слабая отрицательная корреляция между продажами и отзывами пользователей. Зависимость между продажами и мнением критиков есть, умеренная и положительная. Мнение критиков приоритетнее чем мнение пользователей. Это стоит учитывать при сборе отзывов.<br><br>

- Для платформы WiiU наблюдается одинаковая положительная умеренная зависимость между продажами и мнением пользователей и критиков. У платформы X360 - средняя зависимость продаж от мнения критиков, а от мнения пользователей можно сказать ничего не зависит. <br><br>

- Интересно, что продажи игр на платформе Wii напрямую зависят от мнения пользователей, а вот отзывов критиков нет.
Мнение пользователей слабо влияет на продажи, в то время как мнение критиков оказывает некоторое влияние.<br><br>

- Самый популярный жанр - Shooter, а самый не популярный - Adventure. Затем, с уровнем "популярности" вдвое меньшей идут - Sports, Platform, Fighting, Racing.<br><br>

- Мнение пользователей в основном слабо влияет на продажи, в то время как мнение критиков оказывает некоторое влияние. Однако по платформе "WiiU" имеется корреляция продаж от мнения пользователей и она почти такая же, что и корреляция с мнением критиков.
Мнение критиков имеет более высокий приоритет чем мнение пользователей.<br><br>
